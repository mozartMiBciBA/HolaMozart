**Ahora vamos a leer las imagenens, duardarlas en un arreglo y llamarlas para poder vee las imagenes**

// Se utiiza para reconocimiento imagenes en python

```
import cv2 
```

// os se utiliza para poder utilizar los directorios

```
import os
import numpy as np
import matplotlib.pyplot as plt

ruta_tigres= "/content/tigres"
tigres_training= []
img_size=80


for img in os.listdir(ruta_tigres):
  img= cv2.imread(os.path.join(ruta_tigres, img))
  img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img_gray_resize = cv2.resize(img_gray,(img_size, img_size))
  tigres_training.append([img_gray_resize])


print(len(tigres_training))
print(tigres_training)
```

Para ver en una grafica

```
plt.figure
plt.imshow(np.squeeze(tigres_training[0])) # np.squeeze para reducir a un formato de 2 dimensiones
plt.colorbar()
plt.grid(False)
plt.show()
```
